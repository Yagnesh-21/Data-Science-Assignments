###  Dimensionality Reduction with PCA

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
wine_norm=scale(wine1)
```

```python
wine_norm
```

```python
pca=PCA()
pca_values=pca.fit_transform(wine_norm)
```

```python
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
plt.plot(var1,color="red")
```

```python
pca_values[:,0:2]
```

```python
# plot between PCA1 and PCA2 
x = pca_values[:,0]
y = pca_values[:,1]
#z = pca_values[:2:3]
plt.scatter(x,y)
```

```python
pca_data=pd.DataFrame(columns=["X","Y"])
pca_data.X=x
pca_data.Y=y
```

```python
pca_data
```

### Clustering with PCA Data

```python
from sklearn.cluster import KMeans
```

```python
# K-means for pca data
model=KMeans(n_clusters=6).fit(pca_data)
```

```python
pca_data.plot(x="X",y="Y",c=model.labels_,kind="scatter",s=50,cmap=plt.cm.coolwarm_r)
```

```python
import sklearn
sklearn.metrics.silhouette_score(pca_data,model.labels_)
```

### Clustering with Original Data

```python
import warnings
warnings.filterwarnings("ignore")
fig=plt.figure(figsize=(8,4))
WCSS=[]
for i in range(1,11): 
    model=KMeans(n_clusters=i).fit(wine_norm)
    WCSS.append(model.inertia_)
    
plt.plot(range(1,11),WCSS)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model1=KMeans(n_clusters=6).fit(wine_norm)
```

```python
y_means=model1.predict(wine_norm)
```

###  Conclusion and Insights

PCA improves clustering performance: The silhouette score improved significantly after applying PCA, suggesting that dimensionality reduction helps in revealing better-defined clusters.

Dimensionality reduction trade-offs: While PCA simplifies the data, it retains enough variance to improve clustering performance, particularly when high-dimensional data may obscure underlying patterns.

### K-Nearest Neighbours Algorithm 

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = zoo.drop('type', axis=1)  # Features
Y = zoo['type']  # Target variable
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2,random_state=40)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

```python
#Predicting on test data
preds = model.predict(x_test) # predicting on test data set 
```

```python
preds
```

### Classifier's Performance 

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,preds))
```

```python
np.mean(preds==y_test)
```

### Grid Search for Algorithm Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
n_neighbors = np.array(range(1,40))
param_grid = dict(n_neighbors=n_neighbors)
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator=model, param_grid=param_grid)
grid.fit(X, Y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

### Visualizing the decision boundaries 

```python
import warnings
warnings.filterwarnings("ignore")
# choose k between 1 to 41
k_range = range(1, 41)
k_scores = []
# use iteration to caclulator different k in models, then return the average accuracy based on the cross validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=5)
    k_scores.append(scores.mean())
# plot to see clearly
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

### Interview Questions

#### Key Hyperparameters in KNN:

n_neighbors (K): The number of neighbors to consider for classification.

metric: The distance metric used to calculate the distance between points.

weights: The weight function used in prediction (uniform or distance-based).

#### Distance Metrics in KNN:

Euclidean Distance (L2 norm): metric='minkowski' with p=2.

Manhattan Distance (L1 norm): metric='minkowski' with p=1.

Minkowski Distance: Generalized distance metric (when p=1 it's Manhattan; when p=2 it's Euclidean).

Cosine Similarity: Measures the cosine of the angle between two vectors.

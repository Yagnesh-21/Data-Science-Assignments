```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
online=pd.read_excel('Online retail.xlsx')
```

```python
online.fillna(0,inplace=True)
```

```python
online
```

```python
tran=[]
for i in range(0,len(online)):
    tran.append([str(online.values[i,j]) for j in range(0,20) if str(online.values[i,j])!='0'])
```

```python
tran
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Encode transactions using TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(tran).transform(tran)
ap = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
ap
```

```python
# Plot frequent items
ap.sum().to_frame('Frequency').sort_values('Frequency', ascending=False)[:25].plot(kind='bar', figsize=(7, 5), title="Frequent Items")
plt.show()
```

### Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
frequent_itemsets = apriori(ap,min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>3]
```

```python
rules[rules.lift>1]
```

### Analysis and Interpretation

```python
# Sort the rules by lift to identify the strongest relationships
rules = rules.sort_values(by='lift', ascending=False)

# Display the top 10 rules with the highest lift
print(rules.head(10))
```

```python
# Rule interpretation function
def interpret_rules(rules):
    for idx, rule in rules.iterrows():
        antecedent = ', '.join(list(rule['antecedents']))
        consequent = ', '.join(list(rule['consequents']))
        support = rule['support']
        confidence = rule['confidence']
        lift = rule['lift']
        
        print(f"Rule: If a customer buys {antecedent}, they are likely to buy {consequent}.")
        print(f"Support: {support:.2f}, Confidence: {confidence:.2f}, Lift: {lift:.2f}")
        print("-" * 80)

# Apply the function to the top 10 rules
interpret_rules(rules.head(10))
```

### Interview Questions

#### What is Lift and Why is it Important?

Lift measures the strength of a rule by comparing the observed co-occurrence of items (antecedent and consequent) with their expected co-occurrence if they were independent.

Importance: A lift > 1 means the rule is significant and the items are more likely to be bought together than by chance. A lift < 1 suggests negative correlation, and lift = 1 indicates no association.

#### What is Support and Confidence? How Are They Calculated?

Support: The proportion of transactions that contain both the antecedent and consequent. It indicates how frequently the rule occurs in the dataset.

Formula: Support(𝐴→𝐵)=Transactions containing 𝐴 and 𝐵/Total transactions
​
 
Confidence: The proportion of transactions where the antecedent is present and the consequent also occurs. It indicates the rule's reliability.

Formula: Confidence(𝐴→𝐵)=Transactions containing 𝐴 and 𝐵/Transactions containing 𝐴

#### Limitations/Challenges of Association Rule Mining:

Scalability: Processing large datasets can be computationally expensive.

Sparsity: In sparse data, meaningful patterns may be hard to identify.

Meaningful Rules: Many generated rules may not be practically useful, requiring further filtering.

Cold Start Problem: Association rules struggle when there is little historical data (new items or customers).

These challenges require careful tuning of thresholds like support, confidence, and lift.
